```python
import os

# 1) Set your absolute path to the project
BASE_PATH = "/home/jan/studienarbeit/Studienarbeit-CODE_Semantische_Segmentation"

# 2) Construct the path to the "train_hyper_..." folder
train_run_folder = os.path.join(
    BASE_PATH,
    "HyperparameterLOG",
    "lraspp_mobilenet_v3_large",
    "Hyperparameter_Tuning_Deeplabv3",
    "train_hyper_db3db980_37_auto_cast=True,batch_size=8,learning_rate=0.0046,max_epochs=100,weight_decay=0.0000_2025-01-20_02-39-29"
)

# 3) Construct the path to the specific checkpoint folder
checkpoint_folder = os.path.join(
    train_run_folder,
    "checkpoint_000004"
)

def print_folder_contents(folder_path: str):
    """
    Prints the subfolders and files in the given folder path.
    Also checks if a checkpoint.pkl file is found.
    Raises ValueError if the folder does not exist.
    """
    print(f"Inspecting folder:\n  {folder_path}\n")
    
    if not os.path.isdir(folder_path):
        raise ValueError(f"Folder does not exist: {folder_path}")

    # List subfolders
    subfolders = [
        name for name in os.listdir(folder_path)
        if os.path.isdir(os.path.join(folder_path, name))
    ]
    print("Subfolders found:")
    for sf in subfolders:
        print(f"  - {sf}")

    # List files
    files = [
        name for name in os.listdir(folder_path)
        if os.path.isfile(os.path.join(folder_path, name))
    ]
    print("\nFiles found:")
    for f in files:
        print(f"  - {f}")
    
    # Check if checkpoint.pkl is in this folder
    ckp_path = os.path.join(folder_path, "checkpoint.pkl")
    if os.path.isfile(ckp_path):
        print(f"\n[Found checkpoint.pkl at]: {ckp_path}")
    else:
        print("\nNo checkpoint.pkl found in this folder.")

    print("-" * 60 + "\n")

# Now let's print the contents of the two folders:
print_folder_contents(train_run_folder)
print_folder_contents(checkpoint_folder)
```
